```python
import pandas as pd 

data = pd.read_csv("E:\\Weka\\data\\train.csv")
display(data)

```

```python
print(data.isnull().sum().sum()) ## total misising values .
print(data.isnull().sum()) ## total misising values.
```

```python
print(data.iloc[0])
```

```python
data["Cabin"].fillna(data['Cabin'].mode()[0])
display(data)
```

```python
data["Age"].fillna(data['Age'].mean())
display(data)
```

```python
data['SexCode']=pd.Categorical(data['Sex']).codes
data['EmbarkedCode']=pd.Categorical(data['Embarked']).codes
display(data)
```

```python
#feature_cols=['PassengerId','Pclass','Age','SibSp','Parch','Fare','SexCode','EmbarkedCode']
feature_cols=['SexCode','Pclass','Age']
X= data[feature_cols]
display(X)
```

```python
Y=data.Survived
display(Y)
```

```python
from sklearn.model_selection import train_test_split
Xtrain, X_test,Y_train, Y_test  = train_test_split(X,Y,test_size=0.80,random_state=100,shuffle=True)
```

```python
print(Xtrain.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

model = DecisionTreeClassifier()
clf = model.fit(Xtrain,Y_train)
y_pred = model.predict(X_test)

print("Accuracy (Precision):",metrics.accuracy_score(Y_test, y_pred)) 
print('precision score(Precision): %.2f' % metrics.precision_score(Y_test,y_pred))
print('recall score o sensibility(Número total de casos reales positivos reales que se predijeron correctamente): %.2f' % metrics.recall_score(Y_test,y_pred))
print('f1 score (Combina precisión y recuperación): %.2f' % metrics.f1_score(Y_test,y_pred))
```

```python
from sklearn import tree
tree.plot_tree(clf)
```

```python
from sklearn.tree import export_text
r = export_text(clf, feature_names=feature_cols)
print(r)
```

```python
from sklearn.tree import export_graphviz

export_graphviz(clf,out_file="titanic.dot",class_names=feature_cols,feature_names=clf.feature_names_in_,filled=True)
```
