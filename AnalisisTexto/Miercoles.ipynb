## Control de Asistencia

```python
import pandas as pd
import re

matriculados= pd.read_excel("C:\\Users\\User\\Downloads\\ControlAsistencia\\Miercoles\\MatriculadosMier.xlsx") 
matriculados['NombreSinEspacios']=(matriculados['Nombre'].str.replace(" ","*"))
display(matriculados)
```

```python
fecha="28-05-2025"
asistencia= pd.read_csv("C:\\Users\\User\\Downloads\\ControlAsistencia\\Miercoles\\"+str(fecha)+".csv",sep="\t",encoding="UTF-16")
asistencia['Nombre1']=asistencia['Nombre completo'].str.replace(" ","*")
display(asistencia)
```

## Contro asistencia

```python
asistenciaNoRepetida=asistencia["Nombre1"].unique()

asistio = set()
noAsistio=[]
for m in matriculados["NombreSinEspacios"]:    
    if m in asistenciaNoRepetida:
        asistio.add(m) 
    else:
        noAsistio.append(m)   
    

print("\n Asistió.........\n")
for e in asistio:
    print(e)

print("\n------------No Asisitió-------------- ")
for na in noAsistio:
    print(na)
```

```python
dfInasistencia = pd.DataFrame(noAsistio, columns=['Nombre Estudiante'])
dfInasistencia["Fecha"]=fecha

Inasistencia= pd.read_excel("C:\\Users\\User\\Downloads\\ControlAsistencia\\Miercoles\\InasistenciaMiercoles.xlsx") 
frames = [Inasistencia,dfInasistencia]
result = pd.concat(frames)
result.to_excel("C:\\Users\\User\\Downloads\\ControlAsistencia\\Miercoles\\InasistenciaMiercoles.xlsx",index=False)
print(result)
```
