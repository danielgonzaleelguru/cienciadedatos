## Boston Housing

```python
import pandas as pd
import numpy as np


df = pd.read_csv("data\\housing.csv", delimiter=r"\s+",header=None)
print(df.shape)
display(df)

```

```python
print(df.info())
```

```python
df.columns=['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV']
display(df)
```

```python
#df = df.astype({"MEDV": int})
df = df.astype(int)
print(df.info())
```

```python
from sklearn.model_selection import train_test_split


#feature_cols=['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV']
feature_cols=['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']
#feature_cols=['CHAS','RAD']
X= df[feature_cols]
print(X)
Y=df.MEDV
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.75,random_state=123,shuffle=True) 

```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn import metrics

objetoModelo= DecisionTreeRegressor(random_state=0)
modelGenerado=objetoModelo.fit(X_train,Y_train)
y_pred= objetoModelo.predict(X_test)




```

## Arbol de Decisión para el modelo de regresión. Incluye el Coeficiente de determinación o R-Cuadrado.

```python
from sklearn.tree import export_graphviz


export_graphviz(objetoModelo,out_file="BostonHousing.dot",class_names=feature_cols,feature_names=objetoModelo.feature_names_in_,filled=True)
```

##  Extrayendo los MSE del arbol de decisión

```python
extracted_MSEs = objetoModelo.tree_.impurity # The Hidden magic is HERE
for idx, MSE in enumerate(objetoModelo.tree_.impurity):
    print("Node {} has MSE {}".format(idx,MSE))
   
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error as mae

score = objetoModelo.score(X_test, Y_test)
print("Coeficiente de determinación: "+str(score))

print("Error cuadrático medio:"+str(mean_squared_error(Y_test, y_pred)))
error = mae(Y_test, y_pred)
print("Error absoluto medio:"+str(error))
```
