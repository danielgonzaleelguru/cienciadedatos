```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pingouin as pg
from scipy import stats 

url = ('https://raw.githubusercontent.com/JoaquinAmatRodrigo/' +
       'Estadistica-machine-learning-python/master/data/births.csv')
data = pd.read_csv(url,sep=',')
data.head(10)


```

H<sub>0</sub>: No difference between popolation means.<br/>
H<sub>a</sub>: There is difference between population means.


```python
data.groupby('smoke').size()
```

```python
fig, axs = plt.subplots(2,2,figsize=(10,7))
weightSmokers = data.loc[data.smoke=='smoker','weight']
mu, sigma = stats.norm.fit(weightSmokers)

x_hat = np.linspace(min(weightSmokers), max(weightSmokers),num=100)
y_hat = stats.norm.pdf(x_hat, mu,sigma)

axs[0,0].plot(x_hat,y_hat,linewidth=2, label='normal')
axs[0,0].hist(x=weightSmokers,density=True,bins=20,color="#3182bd",alpha=0.5)

axs[0,0].plot(weightSmokers,np.full_like(weightSmokers,-0.01),'|k',markeredgewidth=1)

pg.qqplot(weightSmokers,dist='norm',ax=axs[0,1])
```
