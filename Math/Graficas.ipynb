# Curvas param√©tricas 

## Bibliotecas

```python
import matplotlib.pyplot as plt
import sympy as sp
from sympy.plotting import plot
import numpy as np
import math as m
```

## Representando funciones

```python


ax = plt.figure().add_subplot(projection='3d')

# Prepare arrays x, y, z
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 2
x = r * np.sin(theta)
y = r * np.cos(theta)

ax.plot(x, y, z, label='Curva param√©trica')
ax.legend()

plt.show()
```

```python
x=np.linspace(0,np.pi,100)
y=np.sin(x)
plt.plot(x,y, color="green")
plt.show()
```

```python
t=np.linspace(0,10,30)
x=t-3*np.sin(t)
plt.plot(x,t,color="red")
plt.show()
```

## Gr√°ficas polares

```python
r = np.arange(0, 2, 0.01)
theta = 2 * np.pi * r

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, r)
ax.set_rmax(2)
ax.set_rticks([0.5, 1, 1.5, 2])  
ax.set_rlabel_position(-22.5)  
ax.grid(True)

plt.show()
```

## Representando C√≥nicas

```python
x=np.linspace(-2,2, 2)
x=np.linspace(-5,5, 2)
x,y=np.meshgrid(x,y)
a=.3
plt.contour(x,y,(y**2 -4*a*x),[0],colors='k')
plt.show()
```

## Representado Superficies

```python
X = np.arange(-10, 10, 0.25)
Y = np.arange(-10, 10, 0.25)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure(figsize=(12,6))
ax= fig.add_subplot(projection='3d')
surface=ax.plot_surface(X,Y,Z, cmap="winter")
fig.colorbar(surface)
plt.show()
            
```
