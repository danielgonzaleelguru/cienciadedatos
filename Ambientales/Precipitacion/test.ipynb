```python
    import numpy as np
    from scipy.interpolate import make_interp_spline
    import matplotlib.pyplot as plt

     # Datos de ejemplo
    x = np.array([1, 2, 3, 4, 5])
    y = np.array([1, 3, 2, 5, 4])

     # Crear una interpolación splínica
    spline = make_interp_spline(x, y, k=3)  # k=3 para splines cúbicos

     # Generar valores x más densos para una curva más suave
    x_suave = np.linspace(x.min(), x.max(), 500)

     # Obtener los valores y suavizados
    y_suave = spline(x_suave)

     # Plotear los datos originales y la curva suavizada (opcional)

    plt.plot(x, y, 'o', label='Datos originales')
    plt.plot(x_suave, y_suave, label='Curva suavizada')
    plt.legend()
    plt.show()
```
