# Histogramas de TRI de los Municipios del Magdalena

```python
import rasterio
from matplotlib import pyplot
from rasterio.plot import show_hist
from rasterio.plot import show 
import os
```

```python
directory ="E://ProyectoQGIS//Magdalena2//"
file_list = os.listdir(directory)
for f in file_list:
     if f.endswith("TRI.tif"):          
          src = rasterio.open(directory+f)     
          fig, (axrgb, axhist) = pyplot.subplots(1, 2, figsize=(14,7))
          show(src, ax=axrgb)
          show_hist(src, bins=50, lw=0.0,ax=axhist, stacked=False, alpha=0.3,histtype='stepfilled', title="Histograma "+f, label=['Frecuencia',"DN"])
          axhist.set_xlabel('TRI') 
          axhist.set_ylabel('Frecuencia')   
          axhist.set_title(f)   
        
```
