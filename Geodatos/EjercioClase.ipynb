Climas del Departamento del Antioquia

```python
import geopandas as gpd
import matplotlib.pyplot as plt

pathShapeFile= "data\\shp\\ANTIOQUIA_SUELOS_VF.shp"

shapefile = gpd.read_file(pathShapeFile)

display(shapefile.head(3))
```

```python
print(shapefile.info())
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

# Control del título y los ejes
ax.set_title('Paisajes del Departamento del Amazonas',  pad = 20,  fontdict={'fontsize':20, 'color': '#4873ab'})
ax.set_xlabel('Longitud')
ax.set_ylabel('Latitud')

from mpl_toolkits.axes_grid1 import make_axes_locatable
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.2)

# Mostrar el mapa finalizado
shapefile.plot(column='PAISAJE', legend=True,cmap='Set3',cax = cax,  ax=ax, zorder=5) 
```
